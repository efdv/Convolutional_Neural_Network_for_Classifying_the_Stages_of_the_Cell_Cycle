```python
from sklearn.model_selection import StratifiedKFold
from keras.utils import to_categorical
from keras.callbacks import EarlyStopping
from keras.utils import to_categorical
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import f1_score
from keras.models import load_model
from sklearn.metrics import ConfusionMatrixDisplay 
import numpy as np
#import pandas as pd
import tensorflow as tf

from models import CNN_conf1
from models import CNN_conf2
from models import CNN_conf3
from models import CNN_conf4
from models import CNN_conf5
```

```python
list_models = [CNN_conf1.base_model(), CNN_conf2.base_model(), CNN_conf3.base_model(), CNN_conf4.base_model(), CNN_conf5.base_model()]
```

```python
def normResults(y_pred):
    yp = []
    for row in y_pred:
        valmax = max(row)
        for i in range(len(row)):
            if row[i] == valmax:
                row[i] = 1
            else:
                row[i] = 0
        
        yp.append(row)
    
    return yp
```

```python
"""Paths and data"""

#paths
root = 'E:/OneDrive - Universidad de Guanajuato/EF-Duque-Vazquez-Doctorado/'

#load data
data = np.load(root+"datasets/CellCycle-dataset-efdv/cellCycle.npy")
labels = np.load(root+"datasets/CellCycle-dataset-efdv/labels.npy")

data = data/255.0
```

```python
skfolds = StratifiedKFold(n_splits=5)
```

```python
i = 0
# df_DATA = pd.DataFrame()
# df_DATA['presicion'] = None
# df_DATA['recall'] = None
# df_DATA['f1_score'] = None 

all_results = []
all_weighted_avg = []  
all_presicion  = []
all_recall = []
all_f1_list = []
all_accuracy = []
all_acc = []
all_loss = []
all_val_loss = []
all_val_accuracy = []

for modelo in list_models:
    j = 0
    for train_index, test_index in skfolds.split(data, labels):

        # df = pd.DataFrame()
        # df['name_model'] = None
        # df['Weighted_Avg'] = None
        # df['acc'] = None
        # df['loss'] = None
        # df['val_loss'] = None
        # df['val_accuracy'] = None

        results = []
        weighted_avg = []  
        presicion  = []
        recall = []
        f1_list = []
        accuracy = []
        
        X_train_folds = []
        y_train_folds = []
        X_test_folds  = []
        y_test_folds  = []
        
        
        #divide data
        X_train_folds = data[train_index]
        y_train_folds = labels[train_index]
        X_test_folds  = data[test_index]
        y_test_folds  = labels[test_index]
            
        
        y_train_folds = to_categorical(y_train_folds)
        y_test_folds = to_categorical(y_test_folds)
        
        """ train"""
        #define callback EarlyStopping
        early_stopping = EarlyStopping(monitor='val_loss', patience=10, verbose=1, restore_best_weights=True)
        #fit model
        model = modelo
        history = model.fit(X_train_folds, y_train_folds, validation_data = (X_test_folds, y_test_folds), epochs=200, batch_size=32, callbacks=[early_stopping])
            
        name_model = 'model_'+str(i)+'_'+str(j)
        print("--------------------------------",name_model,"--------------------------------")

        # df['name_model'] = name_model
        model.save(root+'projects/cell_cycle/modelos/'+name_model+'.h5')

        # df['acc'] = history.history['accuracy']
        all_acc.append(history.history['accuracy'])
        print('Accuracy_model_',i,'_',j,':', history.history['accuracy'])
        all_loss.append(history.history['loss'])
        # df['loss'] = history.history['loss']
        print('loss_model_',i,'_',j,':', history.history['loss'])

        all_val_accuracy.append( history.history['val_accuracy'])
        # df['val_accuracy'] = history.history['val_accuracy']
        print('val_accuracy_model_',i,'_',j,':', history.history['val_accuracy'])
        all_val_loss.append(history.history['val_loss'])
        # df['val_loss'] = history.history['val_loss']
        print('val_loss_model_',i,'_',j,':', history.history['val_loss'])
        all_accuracy.append(accuracy)
        accuracy.append(model.evaluate(X_test_folds, y_test_folds))
       
        
            
        """predictions"""
        y_pred = model.predict(X_test_folds)

        ypred_cm = []
        y_test_folds_cm = []
        for i in range(len(y_pred)):
            ypred_cm.append(np.argmax(y_pred[i]))
            y_test_folds_cm.append(np.argmax(y_test_folds[i]))
            
        yp = [] 
        yp = normResults(y_pred)
            
        n_correct = sum(yp == y_test_folds)
        results.append(n_correct/len(y_pred))
        all_results.append(results)
        #df = pd.DataFrame({'accuracy': ';'.join(results)}, index=[j])
        #df['accuracy'] = results
        print('Accuracy_',i,'_',j,':', n_correct/len(y_pred))    
        yp2 = np.array(yp)

        ConfusionMatrixDisplay.from_predictions(y_test_folds_cm,ypred_cm)
            
           
        for ii in range(7):
            presicion.append(precision_score(y_test_folds[:,ii],yp2[:,ii]))
            print('Presicion_',i,'_',j,'_stage_',ii, precision_score(y_test_folds[:,ii],yp2[:,ii]))
            # df_DATA['presicion'] = precision_score(y_test_folds[:,ii],yp2[:,ii])
            recall.append(recall_score(y_test_folds[:,ii],yp2[:,ii]))
            print('Recall: ',i,'_',j,'_stage_',ii, recall_score(y_test_folds[:,ii],yp2[:,ii]))
            # df_DATA['recall'] = recall_score(y_test_folds[:,ii],yp2[:,ii])

            f1 = f1_score(y_test_folds[:,ii],yp2[:,ii])
            print('F1: ',i,'_',j,'_stage_',ii, f1)
            # df_DATA['f1_score'] = f1
            f1_list.append(f1)
        
        all_presicion.append(presicion)
        all_recall.append(recall)
        all_f1_list.append(f1_list)

        #df = pd.DataFrame({'presicion': ';'.join(presicion)}, index=[j])
        #df['presicion'] = presicion
        #df = pd.DataFrame({'recall': ';'.join(recall)}, index=[j])
        #df['recall'] = recall
        #df['f1_score'] = f1_list
        #df = pd.DataFrame({'f1_score': ';'.join(f1_list)}, index=[j])
        
        f1_list = np.array(f1_list)    
        weight = np.array(list(sum(y_test_folds == 1)))
        weighted_sum = np.sum(f1_list * weight)
        total_weight = np.sum(weight)
        weighted_avg.append(weighted_sum / total_weight)
        all_weighted_avg.append(weighted_avg)
        print('Weighted_Avg',i,'_',j,':',(weighted_sum / total_weight))

        # df['Weighted_Avg'] = (weighted_sum / total_weight)

        pathsave = root+'projects/cell_cycle/info/'+name_model+'.xlsx'
        pathsave2 = root+'projects/cell_cycle/info/'+name_model+'other_metrics.xlsx'
        # df.to_excel(pathsave, index=False)
        

        
            
               
        
        print('------------------------------------------------------------------')

        j += 1
    i += 1
    # df_DATA.to_excel(pathsave2, index=False)
```
